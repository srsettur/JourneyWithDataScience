```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

Approach-1, Using normal distribution
Steps: 
1. Creating a normal distribution of random values with given mean and error
2. Meansuring the grade from distribution.
3. Finding the confidence interval with 95% of confidence. 
4. Reported standard deviation is treated as grade error

```python
# given ground and rise data
sampleSize = 100000
groundVal = 2744
riseVal = 466

groundRMS = 4.9
riseRMS = 8.3

np.random.seed(100)

simGroundVal = np.random.normal(groundVal, groundRMS, size=sampleSize)
simriseVal = np.random.normal(riseVal, riseRMS, size=sampleSize)


simError = simriseVal/simGroundVal
```

```python
sns.histplot(simGroundVal)
```

```python
# plt.hist(simriseVal)
sns.histplot(simriseVal)
```

```python
# Computing grade error with 95% of confidence interval
print("{0} +/- {1}".format(simError.mean(), 2*simError.std()))
```

Approach-2, using uncertainties
Steps:
1. Creating ground and rise numbers with uncertainties
2. Calculate the grade value.
3. Obtain the error component using 'error_components'

```python
from numpy import *
from uncertainties import ufloat
```

```python
groundVal = 2744
riseVal = 466

groundRMS = 4.9
riseRMS = 8.3
```

```python
groundComplexVal = ufloat(groundVal, groundRMS)
riseComplexVal = ufloat(riseVal, riseRMS)
```

```python
print(groundComplexVal)
print(riseComplexVal)
```

```python
Grade = riseComplexVal/groundComplexVal
```

```python
print(Grade)
```
